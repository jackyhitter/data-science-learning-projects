```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=10000, n_features=2, n_clusters_per_class = 1, n_redundant = 0, weights = [0.99], random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 1, .1, 0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
class_weight = [{0:w, 1:y} for w in [1,10,50,100] for y in [1,10,50, 100]]
```

```python
class_weight
```

```python
params = dict(penalty = penalty, C = c_values, solver = solver, class_weight = class_weight)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy')
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
score = accuracy_score(y_test, y_pred)
print(f"acc_score : {score}")
cm = confusion_matrix(y_test, y_pred)
print(f"confusion matrix : {cm}")
print(f"{classification_report(y_test, y_pred)}")
```

# ROC Curve and ROC AUC Score

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
X, y = make_classification(n_samples=1000, n_classes=2, random_state=42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.75, random_state=25)
```

```python
dummy_model_prob = [0 for _ in range(len(y_test))]
```

```python
model = LogisticRegression()
model.fit(X_test, y_test)
```

```python
model_prob = model.predict_proba(X_test)
model_prob
```

```python
model_prob = model_prob[:, 1]
```

```python
dummy_model_auc = roc_auc_score(y_test, dummy_model_prob)
model_auc = roc_auc_score(y_test, model_prob)
print(dummy_model_auc)
print(model_auc)
```

```python
dummy_fpr, dummy_tpr, _ = roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr, thresholds = roc_curve(y_test, model_prob)
```

```python
np.mean(thresholds[1:])
```

```python

plt.plot(dummy_fpr, dummy_tpr, linestyle = "--", label = 'Dummy Model')
plt.plot(model_fpr, model_tpr, marker = '.', label = 'Logistic')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,6))

plt.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
plt.plot(model_fpr, model_tpr, marker='.', label='Logistic')

ax = fig.add_subplot(111)

for fpr, tpr, thresh in zip(model_fpr, model_tpr, thresholds):
    ax.annotate(f'{thresh:.2f}', (fpr, tpr))

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
