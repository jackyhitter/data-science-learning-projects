```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0, random_state=25)
```

```python
print(X, y , sep=" \n hi there \n")
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel = 'linear')
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
plt.scatter(pd.DataFrame(X_test)[0], pd.DataFrame(X_test)[1], marker = '*', color = 'blue')
plt.plot(pd.DataFrame(X_test)[0], y_pred, linestyle = '--', color = 'r')
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rbf = SVC(kernel = 'rbf')
rbf.fit(X_train, y_train)
y_pred_rbf = rbf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_rbf))
print(confusion_matrix(y_test, y_pred_rbf))
```

```python
polynomial = SVC(kernel='poly')
polynomial.fit(X_train, y_train)
y_pred_poly = polynomial.predict(X_test)
print(classification_report(y_test, y_pred_poly))
print(confusion_matrix(y_test, y_pred_poly))
```

```python
sigmoid = SVC(kernel = 'sigmoid')
sigmoid.fit(X_train, y_train)
y_pred_sigm = sigmoid.predict(X_test)
print(classification_report(y_test, y_pred_sigm))
print(confusion_matrix(y_test, y_pred_sigm))
```

# Hyperparameter Tunning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C' : [0.1, 1, 10, 100, 1000],
    'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel' : ['rbf']
}
```

```python
grid = GridSearchCV(SVC(), param_grid= param_grid, cv = 5, refit=True, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
y_pred_grid = grid.predict(X_test)
print(classification_report(y_test, y_pred_grid))
print(confusion_matrix(y_test, y_pred_grid))
```
