```python
class myObject:
    def __init__ (self, name):
        self.name = name
        print(f"Object {self.name} created")
    
    def __del__ (self):
        print(f"Object {self.name} deleted")
        
obj1 = myObject("obj1")
obj2 = myObject("obj2")
obj1.ref = obj2
obj2.ref = obj1
```

```python
del obj1
del obj2
```

```python
#  GENERATORS FOR MEMORY EDDICIENCY

def generate_numbers(n):
    for i in range(n):
        yield i

# using generators
for num in generate_numbers(100000):
    print(num)
    if num>10:
        break
```

```python
#  PROFILING MEMORY USAGE WITH TRACEMALLOC

import tracemalloc

def create_list():
    return [i for i in range(10000)]

def main():
    tracemalloc.start()
    create_list()
    
    snapshot = tracemalloc.take_snapshot()
    top_stats = snapshot.statistics('lineno')
    
    print("[Top 10]")
    for stat in top_stats[:10]:
        print(stat)
    
```

```python
main()
```
